# **Import Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.impute import SimpleImputer
from sklearn.metrics import roc_curve, roc_auc_score, classification_report, confusion_matrix
from sklearn.model_selection import learning_curve
from sklearn.model_selection import ShuffleSplit
```

# **Loading data**

```python
df = pd.read_csv("LoanDefault.csv")
print(df)
```

```python
df.head()
```

```python
df.isnull().sum()
```

# **Convert missing values to NaN**

```python
df = df.replace("?", np.nan)
```

# **Convert categorical variables to numeric**

```python
df["REASON_CDE"] = df["REASON_CDE"].astype("category").cat.codes
df["JOB_CDE"] = df["JOB_CDE"].astype("category").cat.codes
```

# **Separate features and target**

```python
X = df.drop("TARGET", axis=1)
y = df["TARGET"]
```


# **Imputation of missing values**

```python
imputer = SimpleImputer(strategy="mean")
X = imputer.fit_transform(X)
```

# **Split data into train and test sets**

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)
```


# **Create and train a LogisticRegression model**

```python
model = LogisticRegression()
model.fit(X_train, y_train)
```

# **Predict the probability of loan default for the test set**

```python
y_pred = model.predict_proba(X_test)[:, 1]
```

# **Evaluate the model's performance**

```python
print("Accuracy:", model.score(X_test, y_test))
print("AUC:", roc_auc_score(y_test, y_pred))
```

# **ROC Curve**

```python
fpr, tpr, thresholds = roc_curve(y_test, y_pred)
plt.plot(fpr, tpr, label="AUC: " + str(roc_auc_score(y_test, y_pred)))
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.show()
```


# **Classification Report**

```python
print(classification_report(y_test, model.predict(X_test)))
```

# **Confusion Matrix**

```python
cm = confusion_matrix(y_test, model.predict(X_test))
plt.imshow(cm, cmap="Blues")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

```python
from sklearn.metrics import precision_recall_curve

precision, recall, thresholds = precision_recall_curve(y_test, y_pred)
plt.plot(recall, precision, label="Precision-Recall Curve")
plt.xlabel("Recall")
plt.ylabel("Precision")
plt.legend()
plt.show()
```
